**About the plot:**

This plot breaks down how many reported crime incidents resulted in an arrest versus those that did not. The dataset includes a boolean field called “Arrest”, where each record indicates whether an arrest was made for that specific incident at the time the report entered the system. A value of True means the police made an arrest connected to that event. A value of False means no arrest was recorded at that point, which could be because the suspect wasn’t identified, the case was still under investigation, or the incident did not lead to an arrest for other reasons.

This visualization is useful because it provides a quick snapshot of how often reported crimes lead to arrests. It highlights the gap between the total number of incidents and the subset that result in immediate enforcement action, offering insight into policing outcomes and the complexity of criminal investigations. (Plotted in the Data Exploration section below.)

# Pushing the entire dataset on Github won't work

GitHub caps individual files at 100 MiB (roughly 104.9 MB), so tossing a 2 GB dataset in there is basically asking for rejection. Since most of the older data (going back to 2001) isn’t essential for what I’m doing, I’m trimming the dataset to include only records from 2020 onward. That slice slashes the storage footprint while still leaving plenty of fresh, high-quality data for analysis and visualizations.

I will store and version-control only this subset in the repo and, if needed, load the full dataset directly from the portal via URL instead of keeping it in Git.

```python
# Create a smaller subset: crashes from 2020 onward, to stay under GitHub's file size limits

data["Date"] = pd.to_datetime(data["Date"], errors = "coerce")
data = data.dropna(subset = ["Date"])
data["Year"] = data["Date"].dt.year

cols = ["Year"] + [c for c in data.columns if c!= "Year"]
data = data[cols]
data = data.sort_values(by = "Year")

data_slc = data[data["Year"] >= 2020].copy()
data_slc.head(5)
```

```python
# Getting information about the sliced dataframe Part 1
data_slc.shape
```

```python
# Column List

data_slc.columns.to_list()
```

```python
# Analysing top 5 Crimes (Not the main plot for Project Part 1)
data_slc["Primary Type"].value_counts().head(5).plot(kind="bar")
```

```python
# Saving the sliced dataset as csv file

data_slc.to_csv("Crimes_-_2020_to_Present.csv", index = False)
```

```python
# Plot (Arrest Rate Breakdown)

import matplotlib.pyplot as plt

arrest_counts = data_slc["Arrest"].value_counts()

plt.figure()
plt.bar(arrest_counts.index.astype(str), arrest_counts.values)
```

## Plot Summary

Summarize the characteristics of the dataset in words: what does it represent, what are the fields/columns/rows, what data types are they, etc.

**Plot (Arrest Rate Breakdown)**

In this dataset, the “Arrest” column is a boolean field that indicates whether an arrest was made for that specific incident at the time the data was recorded.

True means the police did make an arrest in connection with the reported incident.

False means the police did not make an arrest at that point in time, this might be because the suspect wasn’t identified, wasn’t apprehended, the case is still open, or the report did not lead to an arrest for other reasons.

It is important to note that “False” does not necessarily mean no arrest was ever made, only that an arrest had not been logged at the time the incident entered the public dataset. Because of this, the True/False split gives a snapshot of arrest activity tied to crime reports, but it does not represent final case outcomes or convictions.
