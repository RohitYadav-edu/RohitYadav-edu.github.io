```python
real_data = pd.read_csv("/Users/rohityadav/Desktop/Assignments/Sem 3/Data Viz/Project/data/Crimes_-_2001_to_Present.csv", low_memory = False)
real_data.shape
```

```python
real_data.sort_values("Year", inplace = True)
real_data.reset_index(drop = True, inplace = True)
year_2015_start = real_data[real_data["Year"] == 2015].index[0]
year_2025_end = real_data[real_data["Year"] == 2025].index[-1]
```

```python
year_row_info = {}
for year in real_data["Year"].unique():
    row_start = int(real_data[real_data["Year"] == year].index[0])
    row_end = int(real_data[real_data["Year"] == year].index[-1])
    total_rows = row_end - row_start + 1
    year_row_info[int(year)] = {
        "Starting row": row_start, 
        "Ending row": row_end,
        "Total rows": total_rows
    }

for year, info in year_row_info.items():
    print(f"Year{year}: {info}")
```

```python
sliced_real_data = real_data[real_data["Year"] >= 2020]
sliced_real_data = sliced_real_data[["ID","Year"] + [col for col in sliced_real_data.columns if col != "ID" and col != "Year"]]
sliced_real_data.to_csv("/Users/rohityadav/Desktop/Assignments/Sem 3/Data Viz/Project/data/Crimes_-_2020_to_Real_Present_Subset.csv", index = False)
sliced_real_data.shape
```

```python
years = sorted(real_data["Year"].unique())

for i in range(len(years) - 1):
    year1 = years[i]
    year2 = years[i + 1]

    sliced = real_data[(real_data["Year"] >= year1) & (real_data["Year"] < year2)]
    sliced = sliced[["ID","Year"] + [col for col in sliced_real_data.columns if col != "ID" and col != "Year"]]

    filename = f"/Users/rohityadav/Desktop/Assignments/Sem 3/Data Viz/Project/data/year wise/Crimes_{year1}_to_{year2}.csv"
    sliced.to_csv(filename, index = False)

    print(f"Saved: Crimes_{year1}_to_{year2}.csv - {sliced.shape[0]} rows")
```
